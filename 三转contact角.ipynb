```python
import numpy as np
import matplotlib.pyplot as plt

# 
theta_m = np.pi / 2
theta_M = 2 * np.pi / 3

# 
x = np.linspace(-1, 1, 400)
y = np.linspace(-1, 1, 400)
X, Y = np.meshgrid(x, y)

#  hat{theta}
hat_theta = np.arctan2(Y, X)

#  f(x, y)
f = (theta_m + theta_M) / 2 + (theta_M - theta_m) / 2 * np.cos(3 * hat_theta)

# 
plt.figure(figsize=(6, 5))
plt.pcolormesh(X, Y, f, shading='gouraud', cmap='turbo')  # 'turbo'  'twilight' 
plt.colorbar(label=r'$f(x,y)$')

plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
def f(x):
    return -1/9*x**3 + 3/2*x**2 + 13/2*x

def g(x):
    return 15 * np.log2(x + 1)

# x  ( -1)
x = np.linspace(-0.9999, 20, 4000)

# 
plt.figure(figsize=(8,5))
plt.plot(x, f(x), label=r'$f(x) = x(-\frac{1}{9}x^2 + \frac{3}{2}x + \frac{13}{2})$', color='blue')
plt.plot(x, g(x), label=r'$g(x) = 15 \log_2(x+1)$', color='red')

# 
plt.scatter([3,15], [f(3), f(15)], color='black', zorder=5)
plt.text(3, f(3)+2, '(3,30)')
plt.text(15, f(15)+2, '(15,60)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of f(x) and g(x)')
plt.legend()
plt.grid(True)
plt.show()
```
