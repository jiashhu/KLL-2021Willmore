```python
from ngsolve import *
from netgen.csg import *
from netgen.meshing import MeshingStep
from ngsolve.webgui import Draw
```

```python
import netgen.meshing as ngm
```

```python
geo          = CSGeometry()
sphere       = Sphere(Pnt(0,0,0), 1)

geo.Add(sphere)
mesh = Mesh(geo.GenerateMesh(maxh=0.3,perfstepsend=ngm.MeshingStep.MESHSURFACE))
# mesh.Curve(2)
Draw(mesh)
```

```python
fes = H1(mesh, order=2)
```

```python
gfu = GridFunction(fes)
vec = gfu.vec.FV().NumPy()
vec[0] = 1
# vec[10] = 1
gfu.vec.data = BaseVector(vec)
Draw(gfu, mesh, "u")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 节点向量
t = np.array([1, 2, 3, 4, 5])
x = np.linspace(0.5, 5.5, 500)

def N0(i, x):
    return np.where((x >= t[i]) & (x < t[i+1]), 1.0, 0.0)

plt.figure(figsize=(8,2))
for i in range(len(t)-1):
    plt.plot(x, N0(i, x), label=f'N{i},0')
plt.title('0-B spline')
plt.legend()
plt.show()
```

```python
def N1(i, x):
    left = (x - t[i]) / (t[i+1] - t[i]) * N0(i, x)
    right = (t[i+2] - x) / (t[i+2] - t[i+1]) * N0(i+1, x)
    return left + right

plt.figure(figsize=(8,2))
for i in range(len(t)-2):
    plt.plot(x, N1(i, x), label=f'N{i},1')
plt.title('1-B spline')
plt.legend()
plt.show()
```

```python
def N2(i, x):
    left = (x - t[i]) / (t[i+2] - t[i]) * N1(i, x)
    right = (t[i+3] - x) / (t[i+3] - t[i+1]) * N1(i+1, x)
    return left + right

plt.figure(figsize=(8,2))
for i in range(len(t)-3):
    plt.plot(x, N2(i, x), label=f'N{i},2')
plt.title('2-B spline')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline


n = 3
k = 2  # 阶数（order）

# 构造端点重复的节点向量
t_start = [0] * k
t_end = [1] * k
# 中间节点个数 = n - k
t_middle = np.linspace(0, 1, n - k + 2)[1:-1]  # 取中间的点，不含0和1
t = np.concatenate([t_start, t_middle, t_end])
print(t)

# 节点向量
# t = np.linspace(0,1,5)

degree = 3

plt.figure(figsize=(10, 6))

for deg in range(degree+1):
    # 1. 绘制二次B样条基函数（degree=2）
    for i in range(len(t) - deg -1):
        knots = t[i:i+deg+2] 
        basis = BSpline.basis_element(knots)
        x = np.linspace(knots[0], knots[-1], 200)
        y = basis(x)
        plt.plot(x, y, label=f'deg: {deg} {np.round(knots,2)}', linewidth=2)


plt.title('B-spline basis functions on {}'.format(t))
plt.legend(loc='upper right', fontsize=8)
plt.xlabel('x')
plt.ylabel('Basis value')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# 控制点
P = np.array([[0,0], [1,2], [2,1], [3,3], [4,0]])
n = len(P)
k = 3  # 3阶2次

# 节点向量（均匀）
t = np.linspace(0, 1, n + k)
print(t)

# 构造B样条曲线
spl = BSpline(t, P, k-1)  # scipy的degree=k-1
x = np.linspace(t[k-1], t[-k], 100)
curve = spl(x)

plt.plot(P[:,0], P[:,1], 'o--', label='Control Points')
plt.plot(curve[:,0], curve[:,1], label='B-spline Curve')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# 控制点
P = np.array([[0,0], [1,2], [2,1], [3,3], [4,0]])
n = len(P)
k = 3  # 阶数（order），三次B样条

# 构造端点重复的节点向量(总数仍然是n+k)
t_start = [0] * k
t_end = [1] * k
# 中间节点个数 = n - k
t_middle = np.linspace(0, 1, n - k + 2)[1:-1]  # 取中间的点，不含0和1
t = np.concatenate([t_start, t_middle, t_end])

print("Knot vector:", t)

# 构造B样条曲线
spl = BSpline(t, P, k-1)  # degree = k-1
x = np.linspace(t[k-1], t[-k], 100)
curve = spl(x)

plt.plot(P[:,0], P[:,1], 'o--', label='Control Points')
plt.plot(curve[:,0], curve[:,1], label='B-spline Curve')
plt.scatter(P[0,0], P[0,1], color='red', label='First Point')
plt.scatter(P[-1,0], P[-1,1], color='green', label='Last Point')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# 控制点
P = np.array([[0,0], [1,2], [4,0], [3,-2], [0,0], [1,2]])
n = len(P)
k = 3  # 阶数（order），三次B样条

t = np.linspace(0, 1, n+k)

print("Knot vector:", t)

## 绘制knots对应的区间
plt.figure(figsize=(8,2))
for i in range(len(t)-3):
    plt.plot(x, N2(i, x), label=f'N{i},2')

# 在横坐标上标出节点
for knot in t:
    plt.axvline(knot, color='gray', linestyle='--', alpha=0.5)
    plt.scatter(knot, 0, color='red', zorder=5)  # 在x轴上画红点

plt.xticks(t)  # 横坐标刻度设为节点位置
plt.title('2-B spline')
plt.legend()
plt.show()


colors = plt.cm.rainbow(np.linspace(0, 1, len(t)-k))
for i in range(k-1, len(t)-k):
    x_piece = np.linspace(t[i], t[i+1], 20)
    curve_piece = spl(x_piece)
    plt.plot(curve_piece[:,0], curve_piece[:,1], color=colors[i-(k-1)], linewidth=3)


plt.plot(P[:,0], P[:,1], 'o--', label='Control Points')
plt.plot(curve[:,0], curve[:,1], label='B-spline Curve')
plt.scatter(P[0,0], P[0,1], color='red', label='First Point')
plt.scatter(P[-1,0], P[-1,1], color='green', label='Last Point')
plt.legend()
plt.show()
```
